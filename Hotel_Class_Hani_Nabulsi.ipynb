## Predictive Modeling Process:

In the dataframe we have different kind of features: Categorical, boolean and numeric (integer and continous).
A good estimator to estimate a multiclass attribute (OUTPUT) according to most important features in our case 
(integer and boolean) is the DECISION TREE.

It work well with categorical variable but not so well with numeric variables according to the mathematical concepts behind (entropy or information). Entropy is defined as the "state of disorder" of the values of a features according to the value of the attribute to predict. We have max value (1) with max disorder and zero with zero disorder or to a specific value of a features corresponds to the same value for the attribute to predict.

Entropy (or information) = -P x log_base2(P)

Information gain = Entropy at the root - Entropy at the node (considering a path of nodes on the decision tree)
The algorith choose the path (set of attributes) according to the max Information gain.

### FIRST SCENARIO: ALL (congruent) attributes:

In our case we don't apply a rigorous FEATURE SELECTION but we try different scenarios in order to optimize our accuracy.

We consider all (the most logical) attributes: integer values--> CAMERE, SUITE, LETTI, BAGNI ; Boolean values-->
From PRIMA_COLAZIONE to ZONA_STAZIONE_FS.

```python
dfFS= df[["CAMERE",     #We create a dataset whit a subset of attributes form our dataset df
          "SUITE",
          "LETTI",
          "BAGNI",
          "IN_ABITATO",
          "SUL_LAGO",
          "VICINO_ELIPORTO",
          "VICINO_AEREOPORTO",
          "ZONA_CENTRALE",
          "VICINO_IMP_RISALITA",
          "ZONA_PERIFERICA",
          "ZONA_STAZIONE_FS"]]

r_state = 14
X = dfFS.drop(dfFS.columns[-1], axis = 1) # drop the last column
y = dfFS.drop(dfFS.columns[0:-1], axis = 1) # drop all the columns but not the last one

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=r_state) # We split the dataset randomly  
                                                                                # in Training Set : 75% and Test Set 25% 
from sklearn import tree

estimator = tree.DecisionTreeClassifier(criterion="entropy", random_state = r_state) # We inizialize a Decision Tree estimator

estimator.fit(X_train, y_train) # We fit/learn from features and the OUTPUT column (we have a supervised model)

y = y.iloc[:,0].values # get the values of all the rows, column 0, they will be arranged into a one column array

y_train = y_train.iloc[:,0].values
y_test = y_test.iloc[:,0].values

y_predicted_train = estimator.predict(X_train)  #We calculate the accuracy on the Training Set
accuracy_train = np.mean(y_train == y_predicted_train) * 100
print("The accuracy on training set is {0:.1f}%".format(accuracy_train))

y_predicted_test = estimator.predict(X_test)    #We calculate the accuracy on the Test Set
accuracy_test = np.mean(y_test == y_predicted_test) * 100
print("The accuracy on test set is {0:.1f}%".format(accuracy_test))
```
